```python
#Code Snippet 1.1: Generating examples.

import numpy as np

def true_fun(X):
    return np.sin(1.5 * np.pi * X)

np.random.seed(0)
n_samples = 40
X = np.random.rand(n_samples)
X_train = np.sort(X[:n_samples//2])
y_train = true_fun(X_train) + np.random.randn(n_samples//2) * 0.1
X_test = np.sort(X[n_samples//2:])
y_test = true_fun(X_test) + np.random.randn(n_samples//2) * 0.1
```

```python
#Plot both the training and testing examples in a scatter plot and show the true 
#function curve for the range [0, 1).

import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
plt.plot(x, true_fun(x), label="True function")
plt.scatter(X_train, y_train, edgecolor="b", s=20, label="Train Examples")
plt.scatter(X_test, y_test, edgecolor="r", s=20, label="Test Examples")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim((0, 1))
plt.ylim((-1.5, 1.5))

plt.legend(loc="best")
plt.title("Toy Dataset")
plt.show()
```

### Task 4: In the previous example, compute the precision, recall, F1-score. Discuss when accuracy is not the most suitable metric.


```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Predictions on the test set
y_pred = clf.predict(X_test)

# Compute precision, recall, and F1-score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the results
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

Imbalanced Classes: When the classes in the dataset are imbalanced, meaning that one class significantly outnumbers the other, accuracy might not be a reliable metric. For instance, if we have 95% of instances belonging to class A and only 5% to class B, a classifier that predicts all instances as class A would still have a high accuracy of 95%, but it would fail to identify any instance of class B.

Different Costs of Errors: In some scenarios, the cost of making one type of error (e.g., false positives) may be much higher or lower than the cost of making another type of error (e.g., false negatives). Accuracy treats all errors equally, which may not align with the real-world impact of those errors.

Sensitive to Misclassification Types: Accuracy doesn't distinguish between different types of misclassifications. Precision, recall, and F1-score provide more insight into the specific types of errors a classifier is making.

Multiclass Classification: In the case of multiclass classification, accuracy may not provide a complete picture, especially if some classes are more important than others. Precision, recall, and F1-score can be more informative for each class.

In summary, precision, recall, and F1-score offer a more nuanced evaluation of a classifier's performance, especially in situations where accuracy may be misleading or not fully representative of the model's effectiveness.

### Task 6: Draw the decision surface for a set of random forests with different number of trees. Do you notice any pattern?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from matplotlib.colors import ListedColormap

# Function to draw decision surface
def plot_decision_surface(X, y, n_estimators, plot_step=0.02):
    cmap = plt.cm.RdBu
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
    
    plt.figure(figsize=(12, 8))
    
    for n_trees in n_estimators:
        clf = RandomForestClassifier(n_estimators=n_trees, criterion="entropy", max_features="sqrt", max_samples=.8, random_state=0)
        clf.fit(X, y)
        estimator_alpha = 1.0 / len(clf.estimators_)
        
        for tree in clf.estimators_:
            Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
            Z = Z.reshape(xx.shape)
            plt.contourf(xx, yy, Z, alpha=estimator_alpha, cmap=cmap)
    
    # Scatter plot for training samples
    colors = ["r", "b"]
    for c in np.unique(y):
        plt.scatter(X[y == c, 0], X[y == c, 1], edgecolor=colors[c], s=20, label=f"Class {c}")

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Decision Surface for Random Forests with Different Number of Trees")
    plt.legend()
    plt.show()

# Use the function with a range of tree numbers
n_estimators_range = [2]
plot_decision_surface(X_train, y_train, n_estimators_range)
```

```python
n_estimators_range = [5]
plot_decision_surface(X_train, y_train, n_estimators_range)
```

```python
n_estimators_range = [10]
plot_decision_surface(X_train, y_train, n_estimators_range)
```

```python
n_estimators_range = [15]
plot_decision_surface(X_train, y_train, n_estimators_range)
```

As the number of trees increases, the decision surface tends to become smoother and more robust. This is because the aggregation of multiple trees helps reduce overfitting to noise in the training data.
With a smaller number of trees, the decision surface may be more sensitive to individual trees, leading to a more fragmented appearance.

### Task 8: Train an SVM model with rbf kernel and draw its decision boundary. What do you notice?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
from matplotlib.colors import ListedColormap

# Function to draw decision boundary for SVM with RBF kernel
def plot_svm_rbf_decision_boundary(X, y, C=1.0, gamma='scale'):
    cmap = plt.cm.RdBu
    plot_step = 0.02

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))

    clf = svm.SVC(kernel='rbf', C=C, gamma=gamma)
    clf.fit(X, y)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    # Scatter plot for training samples
    colors = ["r", "b"]
    for c in np.unique(y):
        plt.scatter(X[y == c, 0], X[y == c, 1], edgecolor=colors[c], s=20, label=f"Class {c}")

    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"SVM Decision Boundary with RBF Kernel (C: {C}, Gamma: {gamma})")
    plt.legend()
    plt.show()

# Use the function with the SVM model parameters
plot_svm_rbf_decision_boundary(X_train, y_train, C=1.0, gamma='scale')
```

### Task 9: Repeat task 8 but with C equal 0.5, 1, 100, and 1000. What do you notice?


```python
# Values of C to experiment with
C_values = [0.5, 1, 100, 1000]

# Plot decision boundaries for each value of C
for C in C_values:
    plot_svm_rbf_decision_boundary(X_train, y_train, C=C, gamma='scale')
```

As C increases, the regularization strength decreases. Higher C values allow the SVM to fit the training data more closely, potentially leading to a decision boundary that captures intricate details in the data.
Smaller C values increase regularization, promoting a smoother decision boundary that generalizes better to unseen data.
The parameter gamma also plays a crucial role in shaping the decision boundary. Adjusting both C and gamma can impact the SVM's performance and generalization.
